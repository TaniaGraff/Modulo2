1. Hacer las importaciones necesarias para abrir un archivo de tipo "csv" en Pandas.

```python
#Importamos la librería Panda.
import pandas as pd
```


2. Abrir el archivo, creando un Dataframe. Aseguraros de que no incluye la columna "Unnamed: 0".

```python
#Abrimos el archivo csv creando un DataFrame en el que expecifico que el índice de las columnas es 0.
df_videogames= pd.read_csv('videogames.csv', index_col=0)
df_videogames
```

3. Una vez creado el Dataframe, muestra las 10 primeras filas.

```python
#Usamos .head incluyendo el numero 10 entre paréntesis para mostrar las primeras 10 filas.
df_videogames.head(10)
```

```python
#Podemos usar .tail para encontar el último índice...
df_videogames.tail(1)
```

```python
#... O .index para encontrar únicamente el dato que nos interesa.
ultimo_indice = df_videogames.index[-1]
ultimo_indice
```

```python
#Usamos .sample() para mostrar 15 filas aleatorias que cambiarán cada vez que ejecutemos el código.
df_videogames.sample(15)
```

7. Guardar el Dataframe en formato Excel y después crear un Dataframe con el Excel. ¿EL Dataframe es similar al que teníais antes?

```python
#Guardamos el DataFrame como excel.
excel_videogames = df_videogames.to_excel("df_videogames.xlsx")
#Creamos un nuevo DataFrame del excel que parece tener el mismo tipo de información.
df_excel_videogames = pd.read_excel("df_videogames.xlsx", index_col=0)
df_excel_videogames
```

8. Muestra la "info" del Dataframe.

```python
#Para asegurarnos de la similitud de los dos DataFrame imprimimos la información de ambos.
print(df_excel_videogames.info())
print(df_videogames.info())
```

9. ¿Tiene nulos o duplicados?

```python
#Buscamos los valores nulos por columna.
print(df_excel_videogames.isnull().sum())
print('_______________________________________')
#Y Buscamos los valores duplicados
duplicados = df_excel_videogames.duplicated().sum()
print('Los valores duplicados son:', duplicados)
```
