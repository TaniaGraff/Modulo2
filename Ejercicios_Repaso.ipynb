```python
def limpiar_datos_stephen_king(datos_stephen_king, diccionario_stephen_king):
    try:
        for datos in datos_stephen_king["results"]:
            diccionario_stephen_king["titulo"].append(datos["book_title"])
            diccionario_stephen_king["fecha publicación"].append(datos["publication_dt"])
            diccionario_stephen_king["sumario"].append(datos["summary"])
            diccionario_stephen_king["url"].append(datos["url"])
    except TypeError as e:
        print(f"Error: {e}")

    # Convierto listas a conjuntos para eliminar elementos duplicados
    diccionario_stephen_king["titulo"] = list(set(diccionario_stephen_king["titulo"]))
    diccionario_stephen_king["fecha publicación"] = list(set(diccionario_stephen_king["fecha publicación"]))
    diccionario_stephen_king["sumario"] = list(set(diccionario_stephen_king["sumario"]))
    diccionario_stephen_king["url"] = list(set(diccionario_stephen_king["url"]))

    # Verifico y relleno con valores nulos si es necesario para que todas las listas tengan la misma longitud
    max_len = max(len(diccionario_stephen_king["titulo"]), len(diccionario_stephen_king["fecha publicación"]),
                  len(diccionario_stephen_king["sumario"]), len(diccionario_stephen_king["url"]))

    diccionario_stephen_king["titulo"] += [None] * (max_len - len(diccionario_stephen_king["titulo"]))
    diccionario_stephen_king["fecha publicación"] += [None] * (max_len - len(diccionario_stephen_king["fecha publicación"]))
    diccionario_stephen_king["sumario"] += [None] * (max_len - len(diccionario_stephen_king["sumario"]))
    diccionario_stephen_king["url"] += [None] * (max_len - len(diccionario_stephen_king["url"]))

    return diccionario_stephen_king
```

## 2. Scraping de Datos de Libros (Selenium - Book Center):

Utiliza la biblioteca Selenium para automatizar la navegación en la página web de Book Center, una tienda en línea de libros.

Realiza búsquedas de libros en el sitio web y navega por las páginas de resultados.

Extrae información detallada de los libros, como el precio (con y sin descuento si está disponible), editorial, año de edición y número de páginas.

Almacena estos datos en una estructura de datos, como una lista de diccionarios.

```python
driver = webdriver.Chrome()

# Navego a la página de La Casa del Libro
driver.get("https://www.bookcenter.es/")
driver.maximize_window()

#Acepto cookies
sleep(5)
driver.find_element("css selector", "#cookie-accept").click()

#Uso input para hacer la búsqueda 
sleep(5)
eleccion_usuario = input("Qué tipo de libro quieres encontrar en la página de deportes")
#Respondo la categoría de "infantil"
respuesta_usuario = driver.find_element("css selector", '#busqueda > input.search-text').send_keys(eleccion_usuario, Keys.ENTER)

sleep(5)
selector = driver.find_element('css selector', '#main-container > div > div > div.bloque-top > div.ordenar > div > select')
opciones = selector.find_elements("tag name", "option")
opciones
```

```python
#Busco la forma de ordenar todos los libros de infantil.
sleep(5)
for categorias in opciones[0:6]:
    print(categorias.get_attribute('value'))
```

```python
#Elijo que se ordenen por título
titulo = opciones[2].text
titulo
```

```python
#Una vez ordenados, busco los títulos de los 10 primeros libros infantiles.
resultados_búsqueda_titulos = {'título': []}

for titulo in range(1, 11):
    try:
        titulos = driver.find_element('xpath', f'//*[@id="main-container"]/div/div/div[2]/ul/li[{titulo}]/form/dl/dd[@class="title"]/a')
        resultados_búsqueda_titulos['título'].append(titulos.text)
    except NoSuchElementException:
        resultados_búsqueda_titulos['título'].append('No disponible')

resultados_búsqueda_titulos
```

```python
#Busco los autores
resultados_búsqueda_autores = {'autor(es)': []}

for titulo in range(1, 11):
    try:
        autores = driver.find_element('xpath', f'//*[@id="main-container"]/div/div/div[2]/ul/li[{titulo}]/form/dl/dd[@class="creator"]')
        autor_texto = autores.text.split('/')
        autor_texto = [autor.strip() if autor.strip() != '' else 'No disponible' for autor in autor_texto]
        resultados_búsqueda_autores['autor(es)'].append(autor_texto[0])
    except NoSuchElementException:
        resultados_búsqueda_autores['autor(es)'].append('No disponible')

resultados_búsqueda_autores
```

```python
#Busco los precios
resultados_búsqueda_precio = {'precio': []}
for titulo in range(0, 10):

    try:
        precios = driver.find_element('css selector', '#main-container > div > div > div.row > ul > li:nth-child({}) > form > div > p > strong'.format(titulo+1))
        resultados_búsqueda_precio['precio'].append(precios.text)
    except NoSuchElementException:
        resultados_búsqueda_precio['precio'].append('No disponible')
        
resultados_búsqueda_precio
```

```python
#Creo lista urls para iterar sobre los datos de los diez primeros libros que se encuentran en la categoría de infantil.
lista_url = []

elementos_a = driver.find_elements('xpath', '//div[@class="row"]/ul[@class="books grid"]/li[@class="item"]//div[@class="portada"]/div/a')

for a in elementos_a:
    url_libro = a.get_attribute('href')
    lista_url.append(url_libro)

print(lista_url)
```

```python
#Busco dentro de las urls de cada libro, las editoriales
resultados_búsqueda_editorial = {'editorial': []}

for url in lista_url[:10]:
    driver.get(url)
    try:
        editoriales = driver.find_element('xpath', '//*[@id="main-container"]/div/div[1]/div/div[1]/div/dl/dd[1]')
        texto_editorial = editoriales.text
        resultados_búsqueda_editorial['editorial'].append(editoriales.text)
    except NoSuchElementException:
        resultados_búsqueda_editorial['editorial'].append('No disponible')

resultados_búsqueda_editorial
```

```python
#Busco dentro de las urls de cada libro, las páginas
resultados_búsqueda_paginas = {'páginas': []}

for url in lista_url[:11]:
    driver.get(url)
    try:
        paginas = driver.find_element('xpath', '//*[@id="main-container"]/div/div[1]/div/div[1]/div/dl/dd[5]')
        numero_paginas = paginas.text
        if numero_paginas.isdigit():
            resultados_búsqueda_paginas['páginas'].append(int(numero_paginas))
    except NoSuchElementException:
        resultados_búsqueda_paginas['páginas'].append('No disponible')

resultados_búsqueda_paginas
```

```python
#Busco dentro de las urls de cada libro, los años de edición.
resultados_búsqueda_años = {'año': []}

for url in lista_url[:11]:
    driver.get(url)
    try:
        años = driver.find_element('xpath', '//*[@id="main-container"]/div/div[1]/div/div[1]/div/dl/dd[2]')
        año_edicion = años.text
        resultados_búsqueda_años['año'].append(años.text)
    except NoSuchElementException:
        resultados_búsqueda_años['año'].append('No disponible')

resultados_búsqueda_años
```

```python
#Convierto en data frame cada diccionario.
df_titulos = pd.DataFrame(resultados_búsqueda_titulos, columns=['título'])
df_autores = pd.DataFrame(resultados_búsqueda_autores, columns=['autor(es)'])
df_precios = pd.DataFrame(resultados_búsqueda_precio, columns=['precio'])
df_editoriales = pd.DataFrame(resultados_búsqueda_editorial, columns=['editorial'])
df_paginas = pd.DataFrame(resultados_búsqueda_paginas, columns=['páginas'])

#Creo un diccionario con todos los datos
diccionario_completo = {
    'título': df_titulos['título'].tolist(),
    'autor(es)': df_autores['autor(es)'].tolist(),
    'precio': df_precios['precio'].tolist(),
    'editorial': df_editoriales['editorial'].tolist(),
    'páginas': df_paginas['páginas'].tolist()}

#Creo el DataFrame final
df_book_center = pd.DataFrame(diccionario_completo)

df_book_center   
```
